```python
cat_feasts=['purpose']
```

```python
data= pd.get_dummies(loans,columns= cat_feasts, drop_first=True)
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x= data.drop('not.fully.paid',axis=1)
y=data['not.fully.paid']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=101)
```

# Using Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt= DecisionTreeClassifier()
```

```python
dt.fit(x_train, y_train)
```

```python
dpred= dt.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, dpred))
print('\n')
print(classification_report(y_test, dpred))
```

# Using Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf= RandomForestClassifier(n_estimators=300)
```

```python
rf.fit(x_train,y_train)
```

```python
rpred= rf.predict(x_test)
```

```python
print(confusion_matrix(y_test, rpred))
print('\n')
print(classification_report(y_test, rpred))
```
